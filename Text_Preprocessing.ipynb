```python
rows = []

for line in raw_text.strip().split("\n"):
    if "=" in line:
        key, value = line.split("=", 1)
        rows.append([key.strip(), value.strip()])
```

```python
df_slang = pd.DataFrame(rows,columns=["short_form", "full_form"])
df_slang.head()
```

```python
df_slang[df_slang["short_form"] == "AFK"]
```

```python
slang_dict = dict(zip(df_slang["short_form"], df_slang["full_form"]))
```

```python
print(slang_dict["AFK"])
```

```python
def expand_slang(text):
    words = text.split()
    return " ".join(slang_dict.get(word, word) for word in words)
print(expand_slang("BRB I am AFK"))
```

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map
        "\U0001F1E0-\U0001F1FF"  # Flags
        "\U00002700-\U000027BF"
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Example
text = "Hello 😊🚀 World 🌍"
print(remove_emojis(text))
```
