```python
b1=r"\(age\s*\d+\),\s*(.*)"#optimal way
match=re.findall(b1,ch)
match
# \( → matches literal (
# age → matches word "age"
# \s* → optional spaces
# \d+ → one or more digits (age number)
# \) → matches )
# ,\s* → comma and optional spaces
# (.*) → captures birthplace
```

```python
def personal_info(ch):
  name=r"^(.*)\s+is"
  age=r"age.(\d+)"
  birth_p=r"\(age\s*\d+\),\s*(.*)"
  b = r"Born:\s*([^\n]+)"

  return{
      'name':re.findall(name,ch),
      'Age': re.findall(age,ch),
      'birth_Place':re.findall(birth_p,ch),
      'Born':re.findall(b,ch)
  }
```

```python
info=personal_info(ch)
for key, value in info.items():
  print(key,':',value)
```
